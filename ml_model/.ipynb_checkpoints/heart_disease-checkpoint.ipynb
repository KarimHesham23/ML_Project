# Introduction

For this project, we will use Heart Disease Dataset. As a heart disease is a major concern globally, we will use data exploratory analysis and machine learning to create a model that can identify individuals who might develop heart problems

## Description Features
- **age:** ِAge of person
- **sex:** male or female
- **chest pain type:** it is represent different types of chest pain and have four value
    - **Typical Angina:** Chest pain or discomfort caused by reduced blood flow to the heart muscle, often triggered by physical exertion or stress, and relieved by rest or medication.
    - **Atypical Angina:** Chest discomfort similar to angina but with different symptoms or triggers, making it harder to diagnose.
    - **Non-anginal Pain:** Chest discomfort or pain not caused by reduced blood flow to the heart; may be due to musculoskeletal issues, gastrointestinal problems, anxiety, or other non-cardiac causes.
    - **Asymptomatic:** Without symptoms; in the context of chest pain, indicates the absence of chest discomfort or pain, which can occur despite the presence of underlying health conditions
- **Fasting Blood sugar:** The level of glucose (sugar) in the bloodstream after a period of fasting, typically measured after at least 8 hours of not eating or drinking anything other than water. It's an important indicator of metabolic health and can help diagnose.
    - **True:** fasting blood sugar > 120 mg/dl
    - **False:** fasting blood sugar < 120 mg/dl
- **Resting electrocardiogram results:** The graphical representation of the heart's electrical activity recorded while the patient is at rest, providing valuable information about cardiac rhythm, conduction, and potential abnormalities.
    - **Normal:** The resting electrocardiogram (ECG) shows regular cardiac rhythm and waveform patterns, indicating no significant abnormalities in heart function.
    - **Having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV):** The ECG reveals deviations in the ST-T wave segment, such as T wave inversions or ST elevation or depression exceeding 0.05 millivolts (mV), suggesting potential cardiac issues such as ischemia or infarction.
    - **Showing probable or definite left ventricular hypertrophy by Estes' criteria:** Based on Estes' criteria, the ECG indicates probable or definite thickening of the left ventricle wall, known as left ventricular hypertrophy, which can be associated with conditions like hypertension or heart disease.
- **Exercise induced angina:**  Chest pain or discomfort experienced during physical exertion or exercise, typically due to reduced blood flow to the heart muscle, indicating underlying coronary artery disease.
    - **Yes**
    - **No**
- **oldpeak:** The magnitude of ST depression observed on an electrocardiogram (ECG) during exercise stress testing, relative to the baseline ST segment observed at rest, indicating the degree of ischemia and providing valuable information for diagnosing cardiovascular conditions such as coronary artery disease
- **the slope of the peak exercise ST segment:** Describes the direction and pattern of change in the ST segment of an electrocardiogram (ECG) during peak exercise, with possible values including
    - upsloping
    - flat
    - downsloping
- **target :**
    - 1 means heart disease
    - 0 mean Normal

```python
from sklearn.base import BaseEstimator, TransformerMixin
class ZerosCholesterolTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X["cholesterol"] = X["cholesterol"].replace(0, np.nan)
        return X
    def set_output(self,transform=None):
        return self
```

## **Fix outliers**


 i will replace outlier with the max or min number which not outliers 

```python
# detect which feature it contains outliers
for feature in numerical_features:
    
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    IQR = q3-q1
    lower = q1- 1.5*IQR
    upper = q3 + 1.5*IQR
    
    if x_train[(x_train[feature] > upper)].any(axis=None):
        print(feature,"yes")
    else:
        print(feature, "no")
```

```python
for col in numerical_features:
    q1 = x_train_cleaned[col].quantile(0.25)
    q3 = x_train_cleaned[col].quantile(0.75)
    iqr = q3 - q1
    low = q1 - (1.5 * iqr)
    high = q3 + (1.5 * iqr)
    x_train_cleaned.loc[x_train_cleaned[col] < low, col] = low
    x_train_cleaned.loc[x_train_cleaned[col] > high, col] = high
```

class to handle outliers in pipline

```python
class OutlierTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        for col in X.columns:
            q1 = X[col].quantile(0.25)
            q3 = X[col].quantile(0.75)
            iqr = q3 - q1
            low = q1 - (1.5 * iqr)
            high = q3 + (1.5 * iqr)
            X.loc[X[col] < low, col] = low
            X.loc[X[col] > high, col] = high
        return X
    def set_output(self, transform=None):
        return self
```

# modeling

```python
from sklearn.linear_model import LogisticRegression,SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import (accuracy_score,
                             f1_score,
                             roc_auc_score,
                             roc_curve,
                             confusion_matrix,
                             auc
                            )
from sklearn.model_selection import (StratifiedKFold,
                                     cross_val_predict,
                                     cross_val_score,
                                     )
```

## choose our model

```python
class Classifier:
    def __init__(self, model, X, y):
        self.model = model
        self.X = X
        self.y = y
    def __str__(self):
        return self.model.__class__.__name__
        
    def evaluate(self, scoring_metrics=["accuracy", "f1", "roc_auc"], n_splits=5, n_jobs=-1):
        evaluation_results = []
        for metric in scoring_metrics:
            cv = StratifiedKFold(n_splits=n_splits)
            scores = cross_val_score(self.model, self.X, self.y, cv=cv, scoring=metric, n_jobs=n_jobs)
            evaluation_results.append(scores.mean())
        return evaluation_results
            
        return evaluation_results
    def plot_roc_curve(self):
        if hasattr(self.model, "decision_function"):
            y_scores=cross_val_predict(self.model,self.X,self.y,method="decision_function",n_jobs=-1)
        else:
            y_scores=cross_val_predict(self.model,self.X,self.y,method="predict_proba",n_jobs=-1)[:,1]
        fpr, tpr, _ = roc_curve(self.y,y_scores)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
        plt.plot([0, 1], [0, 1], 'k--', label='No Skill')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate (FPR)')
        plt.ylabel('True Positive Rate (TPR)')
        plt.title(f'ROC Curve for {self.model.__class__.__name__}')
        plt.legend(loc="lower right")
```

```python
models=[
    LogisticRegression(random_state=42,max_iter=1000),
    SGDClassifier(random_state=42),
    RandomForestClassifier(random_state=42),
    DecisionTreeClassifier(random_state=42),
    SVC(random_state=42),
    KNeighborsClassifier(),
]
```

```python
scores={"accuracy":[],"f1":[],"roc_auc":[]}
model_objects=[]
for model in models:
    model_clf= Classifier(model, x_train_prepared.values, y_train)
    model_objects.append(model_clf)
    acc, f1, roc_auc = model_clf.evaluate()
    scores["accuracy"].append(acc)
    scores["f1"].append(f1)
    scores["roc_auc"].append(roc_auc)
```

```python
models_scores= pd.DataFrame(scores,index=["LogisticRegression","SGDClassifier","RandomForestClassifier","DecisionTreeClassifier","SVC","KNeighborsClassifier"])
```
